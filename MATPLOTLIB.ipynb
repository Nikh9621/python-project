```python
import matplotlib.pyplot as plt
over=['OVER1','OVER2','OVER3','OVER4','OVER5','OVER6']
runs=[23,4,32,15,10,23]
plt.pie(runs,labels=over,autopct='%1.1f%%',startangle=90,colors=['orange','lightgreen','green','blue','hotpink','grey'])
plt.axis("equal")
plt.title('pie chart')
plt.show()
```
